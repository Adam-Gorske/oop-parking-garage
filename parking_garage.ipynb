# PARKING GARAGE

```python
from datetime import datetime
from IPython.display import clear_output

class ParkingGarage():
    '''
        The ParkingGarage class has a few different methods:
        
        - takeTicket:
            decrease amount of tickets available by 1
            decrease amount of parkingSpaces available by 1
            
        - payForParking:
            takes input and stores it into a varaiable
            if payment variable is not empty (ticket has been paid) -> display "Ticket has been paid. You have 15 mins to leave"
            this will update currentTicket dictionary key paid to true
            if ticket has been paid --> "Thank you, have a nice day"
            if ticket has not been paid --> "Please pay"
            once paid "Thank you, have a nice day"
            update parkingSpaces list to increase by 1
            update tickets list to increase by 1
        
        - checkSpots:
            checks to see how many parking spaces are available
            
            
            
        Expected attributes:
        
        - tickets: list []
        - parkingSpaces: list []
        - currentTicket: dictionary {}     
        
        Other details:
        - 100 spaces in total
        - 1 ticket per space
        - $20 / hr
        - $15 / hr 5 - 12 hours
        - $10 / hr 13+ hours
        
    '''
    
    def __init__(self, tickets, parkingSpaces, currentTicket):
        self.tickets = tickets
        self.parkingSpaces = parkingSpaces
        self.currentTicket = currentTicket
        
    def takeTicket(self):
        if len(self.tickets) == 0:
            print("We're sorry, but we're out of parking spaces! Please wait for someone to leave before admitance.")
        else:
            print(f"Thank you for visiting with us! The current date and time is {datetime.now()}.")
            print("Our hourly rates are as follows:\n$20/hr for first 4 hours\n$15/hr 5-12 hours\n$10/hr 13+ hours")
            del self.tickets[0]
            del self.parkingSpaces[0]
            
            
    def payForParking(self):
        hours_parked = int(input("Please check your receipt to see when you entered. How many hours did you park with us?"))
        count = 0
        if hours_parked <= 4:
            payment = int(input(f"${hours_parked * 20} is your total. Please enter the amount here (without $ sign). "))
            if payment == hours_parked * 20:
                self.currentTicket['current'] = True
                self.tickets.append(1)
                self.parkingSpaces.append(1)
                print("Thank you for parking with us!")
            else:
                int(input("Please enter the correct amount (without $ sign). "))
                self.currentTicket['current'] = True
                self.tickets.append(1)
                self.parkingSpaces.append(1)
                print("Thank you for parking with us!")
                
        elif hours_parked <= 12:
            payment = int(input(f"${hours_parked * 15} is your total. Please enter the amount here (without $ sign). "))
            if payment == hours_parked * 15:
                self.currentTicket['current'] = True
                self.tickets.append(1)
                self.parkingSpaces.append(1)
                print("Thank you for parking with us!")
            else:
                int(input("Please enter the correct amount (without $ sign). "))
                self.currentTicket['current'] = True
                self.tickets.append(1)
                self.parkingSpaces.append(1)
                print("Thank you for parking with us!")
                
        elif hours_parked > 12:
            payment = int(input(f"${hours_parked * 10} is your total. Please enter the amount here (without $ sign). "))
            if payment == hours_parked * 10:
                self.currentTicket['current'] = True
                self.tickets.append(1)
                self.parkingSpaces.append(1)
                print("Thank you for parking with us!")
            else:
                int(input("Please enter the correct amount (without $ sign). "))
                self.currentTicket['current'] = True
                self.tickets.append(1)
                self.parkingSpaces.append(1)
                print("Thank you for parking with us!")
                
    def checkSpots(self):
        print(f"There are {len(self.tickets)} parking spaces available.")
        
        
city_center_parking = ParkingGarage(list(range(100)), list(range(100)), {})       
        
def run():
    while True:
        response = input("Welcome to City Center Parking!\nPlease type one of the following: enter garage / leave garage / check availability / exit program ")
        clear_output()
        
        if response.lower() == 'exit program':
            print("Thank you for visiting City Center Parking!")
            break
            
        elif response.lower() == 'enter garage':
            city_center_parking.takeTicket()
            
        elif response.lower() == 'leave garage':
            city_center_parking.payForParking()
            
        elif response.lower() == 'check availability':
            city_center_parking.checkSpots()
            
            
run()
```
